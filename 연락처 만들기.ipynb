```python
import time

Dictionaly = dict()

while True:
    print("_"*10)
    print('|','1. 목록','|')
    print('|','2. 등록','|')
    print('|','3. 삭제','|')
    print('|','4. 변경','|')
    print('|','0. 종료','|')
    print("_"*10)
    num = input('번호를 입력하세요')
    if num.isdigit() == True:
        
        if num == '1':
            print('현재 목록입니다')
            print(Dictionaly)
            break
        elif num == '2':
            print('"등록"란 입니다')
            name = input('단어를 입력해 주세요')
            mean = input('단어의 의미를 입력해주세요')
            Dictionaly['name']=mean
            print('등록 되었습니다.')
            print('작업을 더 하시겠습니까? (y/n)')
            
            while True:
                YN = input('입력란').lower()
                if YN.isalpha() == True and YN in 'y':
                    print('어떤 작업을 하시겠습니까?')
                    print('더 이상가면 겁나 길어진다.')
                    break
                elif YN.isalpha() == True and YN in 'n':
                    print('n을 입력 하였습니다. \n마치겠습니다')
                    break
                else:
                    print('알맞게 써넣어 주세요 \n (y/n)')
            break
                                    
        elif num == '3':
            print('"삭제"란입니다 \n삭제할 이름을 입력해주세요' )
            print('목록입니다',\
                  Dictionaly)
            name = input('이름:')
            try:
                del Dictionaly[name]
                print('.')
                time.sleep(0.5)
                print('.')
                time.sleep(0.5)
                print('.')
                time.sleep(0.5)
                print('.')
                time.sleep(0.5)
                print('.')
                time.sleep(0.5)
                print('삭제 완료')
                time.sleep(1)
                print('작업을 더 하시겠습니까? (y/n)')
                while True:
                    YN = input('입력란').lower()
                    if YN.isalpha() == True and YN in 'y':
                        print('어떤 작업을 하시겠습니까?')
                        print('더 이상가면 겁나 길어진다.')
                        break
                    elif YN.isalpha() == True and YN in 'n':
                        print('n을 입력 하였습니다. \n마치겠습니다')
                        break
                    else:
                        print('알맞게 써넣어 주세요 \n (y/n)')
                break
            except:
                print('이름이 없습니다.\n처음부터 다시해주십시오')
        
        elif num == '4':
            print('"변경"입력란입니다.')
            print('무엇을 변경하시겠습니까?',Dictionaly)
            name = input('바꿀 이름')
            mean = input('바꿀 의미')
            Dictionaly[name]=mean
            print('바뀌엇습니다.\n 감사합니다')
            
            
        else:
            print('0번 입력 되었습니다.\n종료합니다.')
            break
    elif num.isdigit() == False:
        print('번호가 아닙니다.다시 입력하세요')        
    break
```
