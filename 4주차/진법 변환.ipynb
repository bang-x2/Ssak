B진법 수 N이 주어진다. 이 수를 10진법으로 바꿔 출력하는 프로그램을 작성하시오.  

10진법을 넘어가는 진법은 숫자로 표시할 수 없는 자리가 있다. 이런 경우에는 다음과 같이 알파벳 대문자를 사용한다.  

A: 10, B: 11, ..., F: 15, ..., Y: 34, Z: 35  

첫째 줄에 N과 B가 주어진다. (2 ≤ B ≤ 36)  

B진법 수 N을 10진법으로 바꾸면, 항상 10억보다 작거나 같다.  

# 입력  
ZZZZZ 36

# 출력
60466175

```python
def notation2(self):
    s=[]
    alp=string.digits+string.ascii_lowercase
    while len(s) < 5:
        a,b = self//36,self%36
        s.append(b)
        self = a
    return "".join([alp[i] for i in s])
```

```python
notation2(60466175)
```

```python
import string
a = 'zzzzz'
alp=string.digits+string.ascii_lowercase
sum_ = 0
for i,v in enumerate(list(a)):
    val = alp.index(v)
    value=((36**i)*val)
    sum_ += value
print(sum_)
```
