진법이란?

수를 셀 때 자릿수가 올라가는 단위를 기준으로 하는 셈법의 총칭  
사용하는 숫자의 개수가 진법의 숫자를 의미  

사용이유  
CPU가 사용하기 때문  


# 10진법 -> 2진법

10을 2로나눈다


```python
#10진법에서 2진법으로 변환해준다.
bin(10)
```

# 10진법 ->8진법

95를 8로 나누어준다

```python
# 8진법 함수
oct(95)
```

# 10진법 -> 16진법
10개의 수와 알파벳6개를 사용한다.

350을 16으로 나누어준다.

```python
#16진법 변환
hex(350)
```

```python
#진법변환
int(0x15e)
```
